```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA, FastICA
from sklearn.random_projection import GaussianRandomProjection
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping
from scikeras.wrappers import KerasClassifier
import matplotlib.pyplot as plt
```

```python

# Load and preprocess the dataset
data = load_breast_cancer()
X = data.data
y = data.target

# Standardize the dataset
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split the dataset
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python

# Function to create the neural network model
def create_model(learning_rate=0.01):
    model = Sequential()
    print(X_train.shape[1])
    model.add(Dense(8, input_dim=X_train.shape[1], activation='relu'))
    model.add(Dense(4, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    optimizer = Adam(learning_rate=learning_rate)
    model.compile(loss='binary_crossentropy', optimizer=optimizer, metrics=['accuracy'])
    return model

# Create the KerasClassifier
model = KerasClassifier(model=create_model, verbose=0)

# Define the grid search parameters
param_grid = {
    'model__learning_rate': [0.001, 0.01, 0.1],
    'batch_size': [10, 20, 50],
    'epochs': [50, 100, 200]
}
```

```python

# Create the GridSearchCV
grid = GridSearchCV(estimator=model, param_grid=param_grid, n_jobs=-1, cv=3)

# Fit and optimize model on the original dataset
grid_result = grid.fit(X_train, y_train, callbacks=[EarlyStopping(monitor='loss', patience=3)])
print(f"Best: {grid_result.best_score_} using {grid_result.best_params_}")

# Plot learning curves for the original dataset
plt.plot(grid_result.cv_results_['mean_test_score'])
plt.xlabel('Hyperparameter combination index')
plt.ylabel('Accuracy')
plt.title('Learning Curve for Original Dataset')
plt.show()

```

```python
from sklearn.metrics import accuracy_score

best_model = grid_result.best_estimator_

# Fit the best model on the training data (though it should already be fitted)
best_model.fit(X_train, y_train)

# Predict on the test data
y_pred = best_model.predict(X_test)

# Calculate the accuracy
test_accuracy = accuracy_score(y_test, y_pred)
print(f"Test Accuracy: {test_accuracy}")
```

```python

# Perform PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
X_train_pca, X_test_pca, y_train_pca, y_test_pca = train_test_split(X_pca, y, test_size=0.2, random_state=42)

# Function to create the neural network model
def create_model(learning_rate=0.01):
    model = Sequential()
    model.add(Dense(8, input_dim=X_train_pca.shape[1], activation='relu'))
    model.add(Dense(4, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    optimizer = Adam(learning_rate=learning_rate)
    model.compile(loss='binary_crossentropy', optimizer=optimizer, metrics=['accuracy'])
    return model

# Create the KerasClassifier
model = KerasClassifier(model=create_model, verbose=0)
# Create the GridSearchCV
grid = GridSearchCV(estimator=model, param_grid=param_grid, n_jobs=-1, cv=3)
# Define the grid search parameters
param_grid = {
    'model__learning_rate': [0.001, 0.01, 0.1],
    'batch_size': [10, 20, 50],
    'epochs': [50, 100, 200]
}
# Fit and optimize model on the PCA reduced dataset
grid_result_pca = grid.fit(X_train_pca, y_train_pca, callbacks=[EarlyStopping(monitor='loss', patience=3)])
print(f"Best: {grid_result_pca.best_score_} using {grid_result_pca.best_params_}")

# Plot learning curves for the PCA reduced dataset
plt.plot(grid_result_pca.cv_results_['mean_test_score'])
plt.xlabel('Hyperparameter combination index')
plt.ylabel('Accuracy')
plt.title('Learning Curve for PCA Reduced Dataset')
plt.show()
```

```python

# Perform ICA
ica = FastICA(n_components=2)
X_ica = ica.fit_transform(X_scaled)
X_train_ica, X_test_ica, y_train_ica, y_test_ica = train_test_split(X_ica, y, test_size=0.2, random_state=42)

# Fit and optimize model on the ICA reduced dataset
grid_result_ica = grid.fit(X_train_ica, y_train_ica, callbacks=[EarlyStopping(monitor='loss', patience=3)])
print(f"Best: {grid_result_ica.best_score_} using {grid_result_ica.best_params_}")

# Plot learning curves for the ICA reduced dataset
plt.plot(grid_result_ica.cv_results_['mean_test_score'])
plt.xlabel('Hyperparameter combination index')
plt.ylabel('Accuracy')
plt.title('Learning Curve for ICA Reduced Dataset')
plt.show()
```

```python

# Perform RP
rp = GaussianRandomProjection(n_components=2)
X_rp = rp.fit_transform(X_scaled)
X_train_rp, X_test_rp, y_train_rp, y_test_rp = train_test_split(X_rp, y, test_size=0.2, random_state=42)

# Fit and optimize model on the RP reduced dataset
grid_result_rp = grid.fit(X_train_rp, y_train_rp, callbacks=[EarlyStopping(monitor='loss', patience=3)])
print(f"Best: {grid_result_rp.best_score_} using {grid_result_rp.best_params_}")

# Plot learning curves for the RP reduced dataset
plt.plot(grid_result_rp.cv_results_['mean_test_score'])
plt.xlabel('Hyperparameter combination index')
plt.ylabel('Accuracy')
plt.title('Learning Curve for RP Reduced Dataset')
plt.show()
```

```python

# Evaluate each model on the test set
original_score = grid_result.score(X_test, y_test)
pca_score = grid_result_pca.score(X_test_pca, y_test_pca)
ica_score = grid_result_ica.score(X_test_ica, y_test_ica)
rp_score = grid_result_rp.score(X_test_rp, y_test_rp)

# Print the results
print(f"Original Dataset Accuracy: {original_score}")
print(f"PCA Reduced Dataset Accuracy: {pca_score}")
print(f"ICA Reduced Dataset Accuracy: {ica_score}")
print(f"RP Reduced Dataset Accuracy: {rp_score}")

# Compare and justify
best_model = max([original_score, pca_score, ica_score, rp_score])
if best_model == original_score:
    best_method = "Original"
elif best_model == pca_score:
    best_method = "PCA"
elif best_model == ica_score:
    best_method = "ICA"
else:
    best_method = "RP"

print(f"The best performing method is {best_method} with an accuracy of {best_model}")

```
